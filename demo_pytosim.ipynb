# Pytosim example
Requires Python>=3.7

```python
import cytosim
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_fibers(fibers):
    plt.figure()
    for f in fibers:
        p = f.points()
        plt.plot(p[:,0],p[:,1])
    plt.axis('equal')
    plt.show()
```

# Aster simulation
Here we simulate an aster with 32 microtubules of fixed size, larger than the cell radius.

```python
sim = cytosim.start("cym/aster.cym")
```

```python
frame = sim.frame()
plot_fibers(frame["microtubule"])
```

```python
for t in range(100):
    sim.step()
    sim.solve()
frame = sim.frame()
plot_fibers(frame["microtubule"])
```

```python
print(len(frame["microtubule"]))
```
